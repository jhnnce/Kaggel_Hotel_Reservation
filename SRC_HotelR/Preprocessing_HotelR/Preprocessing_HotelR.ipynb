VISUALIZACIÓN DE DATOS

La primera columna innecesaria es 'Booking_ID', que son números que identifican al cliente en el Hotel, no es útil en el modelo, así que procederemos a eliminarlo.

También es innecesaria la columna "arrival_year"

Cambiaremos los valores en la columna 'booking_status', nuestra variable dependiente, los valores Not_Canceled por 1 y los Canceled por 0.

```python
df.hist(figsize=(15,15));
```

```python
df = df.drop(["Booking_ID", "arrival_year"], axis=1)
```

```python
df.booking_status = df.booking_status.replace("Not_Canceled", 1).replace("Canceled", 0)
```

```python
df.booking_status.value_counts()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,10))
sns.heatmap(corr, annot=True);
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(5,5))
sns.heatmap(pd.DataFrame(corr.booking_status), annot=True, cmap="Blues")
```

COLUMNAS MÁS SIGNIFICATIVAS

Las columnas más determinantes y más relacionadas con booking_status, ordenadas en forma descendente con respecto a su relación, son:

lead_time: Mientras un cliente haga una reservación con una fecha mas cercana, es más probable que no cancele su reservación.

    Observamos que el promedio de días es 40, entre los clientes que hacen una reservación sin cancelarlos posteriormente.

no_of_special_requests: Mientras un cliente haga mas pedidos especiales, es más probable que no cancele su reservación.

avg_price_per_room: Mientras menor sea el costo de su habitación es mas probable que el cliente no cancele su reservación.

    Observamos que el precio promedio de las habitaciones es $95 en reservaciones que no fueron canceladas posteriormente.

repeated_guest: Si un cliente es repetido es menos probable que cancele una reservación.

### lead_time

```python
sns.boxplot(x="booking_status", y = "lead_time", data=df);
```

```python
sns.histplot(x="lead_time",hue="booking_status",data=df);
```

### no_of_special_requests

```python
plt.figure(figsize=(17,6))
n=len(df.no_of_special_requests.unique())
for i in range(n):  
    plt.subplot(1,n,i+1)
    plt.ylim(0, 14000)
    plt.yticks(range(0, 12001, 2000))
    ax = sns.countplot(x=df[df["no_of_special_requests"] == i].no_of_special_requests,  hue="booking_status", palette="tab10", data=df)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2,
                height + 0.1,
                '{:.2%}'.format(height / df['no_of_special_requests'].value_counts()[i]),
                ha="center", va="bottom")
plt.tight_layout()
plt.show()
```

### repeated_guest

```python
plt.figure(figsize=(9,6))
n=len(df.repeated_guest.unique())
for i in range(n):
    plt.subplot(1,2,i+1)
    plt.ylim(0, 27000)
    plt.yticks(range(0, 25001, 5000))
    ax = sns.countplot(x=df[df["repeated_guest"] == i].repeated_guest,  hue="booking_status", palette="tab10", data=df)   
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2,
                height + 0.1,
                '{:.2%}'.format(height / df['repeated_guest'].value_counts()[i]),
                ha="center", va="bottom")

plt.show()


```
