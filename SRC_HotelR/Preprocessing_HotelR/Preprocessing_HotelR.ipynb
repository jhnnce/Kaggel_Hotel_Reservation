VISUALIZACIÓN DE DATOS

La primera columna innecesaria es 'Booking_ID', que son números que identifican al cliente en el Hotel, no es útil en el modelo, así que procederemos a eliminarlo.

Tambien es inncesaria la columna "arrival_year"

Cambiaremos los valores en la columna 'booking_status', nuestra variable dependiente, los valores Not_Canceled por 1 y los Canceled por 0

```python
df.hist(figsize=(15,15));
```

```python
df = df.drop(["Booking_ID", "arrival_year"], axis=1)
```

```python
df.booking_status = df.booking_status.replace("Not_Canceled", 1).replace("Canceled", 0)
```

```python
df.booking_status.value_counts()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(15,15))
sns.heatmap(corr, annot=True);
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(5,8))
sns.heatmap(pd.DataFrame(corr.booking_status), annot=True, cmap="Blues")
```

lead_time: Mientras menos dias exista entre la fecha de reserva y la fecha de llegada más probable es que la reserva no se cancele

no_of_special_requests: Mientras un cliente haga mas pedidos especiales en su reservacion es mas probable que no cancele su reservacion.

avg_price_per_room: Es mas probable que un cliente no cancele la reservacion si el precio de su habitacion sea menor.

repeated_guest: Si un cliente es repetido es menos probable que cancele una reservacion.

Meal Plan 1: Si el cliente elige este tipo de plan de comidas en su reserva es mas probable que no cancele su reserva

Room_Type 1, Room_Type 4: Si el cliente hace reservas con este tipo de habitaciones, es mas probable que no cancele su reserva.

Offline, Online, Corporate: Si el cliente hace su reserva por estos medios, es mas probable que no cancele su reserva.
